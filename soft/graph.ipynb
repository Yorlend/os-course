```python
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.figsize'] = (10, 8)
```

```python
data = [
    (16, (6600472, 6501990, 6432315, 6591485, 6323688)),
    (64, (1755067, 1786833, 1635228, 1720914, 1811774)),
    (256, (492662, 532219, 552496, 580939, 783898)),
    (1024, (186367, 207639, 264803, 260969, 196995)),
    (4096, (120247, 142503, 139143, 116871, 114071)),
    (16384, (94942, 105696, 100153, 117702, 100485)),
    (65536, (105536, 105345, 108534, 97992, 108791)),
    (131072, (118275, 81065, 83325, 88661, 91463)),
]
```

```python
plt.yscale('log')
plt.xscale('log')
plt.xlabel('размер буфера, байт')
plt.ylabel('время обработки запроса, нс')
plt.grid(True, 'both')
plt.plot([16, 64, 256, 1024, 4096, 16384, 65536, 131072], [sum(x[1])/5 for x in data], 'r')
```
